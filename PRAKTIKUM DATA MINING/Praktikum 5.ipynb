```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')

df.head()
```

```python
#Total Missing Value per attributes

df.isnull().sum()
```

```python
df1 = df.dropna()

df1
```

```python
df
```

```python
#Fill missing value with mean >>> fillna(.mean())

df2 = df
df['Pendapatan'] = df['Pendapatan'].fillna(df['Pendapatan'].mean())
df['Umur'] = df['Umur'].fillna(df['Umur'].mean())
df2.round({'Pendapatan':1, 'Umur':1})
df2
```

```python
#Round the float values, then convert Float to Int

df2['Pendapatan']=df2['Pendapatan'].round().astype(int)
df2['Umur']=df2['Umur'].round().astype(int)
df2
```
